# R. Kutta (segundo orden)

```python
def R_Kutta(ec,p_ini,x,h=0.1):
   
    tiempos = arange(x[0],x[1]+h,h)   
    sol=zeros((len(tiempos), len(p_ini)) )
    sol[0,:]=p_ini
    for i in xrange(len(tiempos)-1):
        k=h*ec(sol[i,:],tiempos[i])
        f=ec(sol[i,:] +0.5*k, tiempos[i]+0.5*h) ## por la forma en que definimos ponemos primero a la variable dependiente y despues la independiente
        sol[i+1,:] = sol[i,:]+h*f  
    
    return tiempos,sol
```

```python
def p(y,t):
    return array([y[1],-2*y[1]-0.5*sin(y[0])+ cos(t)])
```

```python
t,y= R_Kutta(p,[1,0],[0,100],0.1)
plot(y[:,0],y[:,1])
```

Este metodo es mucho mas facil de usar que " Euler a 2° orden" pero no es util si la función tiene singularidades
